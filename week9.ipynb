```python
import numpy as np


def sigmoid(x):
    return 1 / (1 + np.exp(-x))


v = np.array([[0.1, 0.2, -0.5], [0.3, 0.2, -0.3]])
x = np.array([0, 0, 1])
z1 = np.dot(v, x)
h11 = sigmoid(z1[0])
h12 = sigmoid(z1[1])

print(h11, h12)

w2 = np.array([0.1, 0.2])
z2 = np.dot(w2, np.array([h11, h12]))
y = sigmoid(z2)
print(z2)
print(y)
```
